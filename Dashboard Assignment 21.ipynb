Question no.7 :- Create a date object of your day of birth.

Answer no.7

```python
day_of_birth = '10/11/1993'
day_of_birth_object = datetime.strptime(day_of_birth,"%d/%m/%Y")
print(day_of_birth_object)
```

Question no.8 :- What day of the week was your day of birth?

Answer no.8

```python
week_day = day_of_birth_object.strftime("%A")
print(week_day)
```

Question no.9 :- When will you be (or when were you) 10,000 days old?

Answer no.9

```python
import datetime

start_date = datetime.datetime(year=1993, month=11, day=10)

end_date = start_date + datetime.timedelta(days=10000)

print(end_date.date())
```
