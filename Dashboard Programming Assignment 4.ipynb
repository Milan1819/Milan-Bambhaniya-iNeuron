Question no.1 :- Write a Python Program to Find the Factorial of a Number?

Answer no.1

```python
f = int(input("Enter number you want: "))

fac = 1

if f < 0:
    print("The Factorial of negative numbers doesn't exist.")

elif f == 0:
    print("The Factorial of 0 is 1")

else:
    for i in range(1,f+1): 
        fac = fac*i
    print(f"The Factorial of {f} is: {fac}")
```

Question no.3 :- Write a Python Program to Print the Fibonacci sequence?

Answer no.3

```python
n_terms = int(input("How many terms the user wants to print: "))  
   
n_1 = 0  
n_2 = 1  
count = 0  
  
if n_terms <= 0:  
    print ("Please enter a positive integer, the given number is not valid")  

elif n_terms == 1:  
    print ("The Fibonacci sequence of the numbers up to", n_terms, ": ")  
    print(n_1)  

else:  
    print ("The fibonacci sequence of the numbers is:")  
    while count < n_terms:  
        print(n_1)  
        n_3 = n_1 + n_2  
        n_1 = n_2  
        n_2 = n_3  
        count = count+1
```

Question no.4 :- Write a Python Program to Check Armstrong Number?

Answer no.4

```python
n = int(input("Enter number to be checked: ")) 

order = len(str(n))
sum = 0
T = n

while T > 0:
   digit = T % 10
   sum += digit ** order
   T //= 10

if n == sum:
   print(f"{n} is an Armstrong number")
else:
   print(f"{n} is not an Armstrong number")
```

Question no.5 :- Write a Python Program to Find Armstrong Number in an Interval?

Answer no.5

```python
rs = int(input("Enter interval start: ")) 
re = int(input("Enter interval end: "))

for number in range(rs, re+1):

   order = len(str(number))
    
   sum = 0

   t = number
   while t > 0:
       digit = t % 10
       sum += digit ** order
       t //= 10

   if number == sum:
       print(number)
```

Question no.6 :- Write a Python Program to Find the Sum of Natural Numbers?

Answer no.6

```python
X = int(input("Enter a number to get sum upto it: "))  
  
if X < 0:  
   print("Enter a positive number")  

else:  
   sum = 0  
     
   while(X > 0):  
       sum += X 
       X -= 1  
   print(f"The sum is: {sum}")
```
