Question no.1 :- Write a Python program to check if the given number is a Disarium Number?

Answer no.1

```python
def Disarium(num):
    sum = 0
    while(num>0):
        l= len(str(num))
        n = num%10
        sum= sum + n**l   
        l-=1
        num = num//10
    return sum

num = int(input("Enter the number :"))


sum = Disarium(num)
if sum == num:
    print(num,"is a Disarium number")
else:
    print(num,"is not a Disarium number")
```

Question no.2 :- Write a Python program to print all disarium numbers between 1 to 100?

Answer no.2

```python
def disarium(num):
    sum = 0
    while(num>0):
        l= len(str(num))
        n = num%10
        sum= sum + n**l   
        l-=1
        num = num//10
    return sum

disnums = []

for i in range(1,101):
    sum = 0
    sum = disarium(i)
    if sum == i:
        disnums.append(i)

print("Disarium numbers in range 1 to 100 are:")
print(disnums)
```

Question no.3 :- Write a Python program to check if the given number is Happy Number?

Answer no.3

```python
def Happy(num):
    sum = 0
    while(num>0):
        dig = num%10
        sum= sum + dig**2        
        num = num//10
    return sum

num = int(input("Enter the number to check:"))
res = num


while (res != 1 and res != 4):
    res = Happy(res)

if res == 1:
    print(num,"Is a Happy Number")
else:
    print(num," Is a Unhappy Number")
```

Question no.4 :- Write a Python program to print all happy numbers between 1 and 100?

Answer no.4

```python
def isHappyNumber(num):    
    rem = sum = 0    
              
    while(num > 0):    
        rem = num%10    
        sum = sum + (rem*rem)    
        num = num//10    
    return sum    

print("List of Happy numbers between 1 and 100: ")                

for i in range(1, 101):    
    result = i    
           
    while(result != 1 and result != 4):    
        result = isHappyNumber(result)   
        
    if(result == 1):    
        print(i,end=" ")
```

Question no.5 :- Write a Python program to determine whether the given number is a Harshad Number?

Answer no.5

```python
def Harshad(n):
    sum = 0
    var = n

    while var > 0 :
        sum = sum + var % 10
        var = var // 10    
    return n % sum == 0
 
h = int(input("Enter the number to check: ")) 

if(Harshad(h)): 
    print(f"{h} is a Harshad number")

else: 
    print (f"{h} is not a Harshad number")
```

Question no.6 :- Write a Python program to print all pronic numbers between 1 and 100?

Answer no.6

```python
def Pronic(num):    
    flag = False   
        
    for j in range(1, num+1):               
        if((j*(j+1)) == num):    
            flag = True   
            break   
    return flag   
        
print("Pronic numbers between 1 and 100: ")    
for i in range(1, 101):    
    if(Pronic(i)):    
        print(i,end =" ")
```
