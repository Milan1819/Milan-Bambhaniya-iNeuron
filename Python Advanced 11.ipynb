Question no.1 :- What is the concept of a metaclass?

Answer no.1

~ In object-oriented programming, a metaclass is a class whose instances are classes. Just as an ordinary class defines the behavior of certain objects, a metaclass defines the behavior of certain classes and their instances.

Question no.2 :- What is the best way to declare a class's metaclass?

Answer no.2

~ Python provides the functionality to create custom metaclasses by using the keyword type. Type is a metaclass whose instances are classes. Any class created in python is an instance of type metaclass.

Question no.3 :- How do class decorators overlap with metaclasses for handling classes?

Answer no.3

~ The class decorators overlap sometimes with metaclasses for handling classes because of similarity in their functionality. Although they are typically used for managing or augmenting instances, class decorators can also augment classes, independent of any created instances.

Question no.4 :- How do class decorators overlap with metaclasses for handling instances?

Answer no.4

~ Decorators overlap with metaclasses for handling instances because decorators can be used to manage both instances and classes, and they intersect with metaclasses in the second of these roles.
