Question no.1 :- What is the result of the code, and why?

```python
def func(a, b=6, c=8):
    print(a, b, c)
func(1, 2)
```

Answer no.1

~ The value of a is 1,it is assigned at the time of function call. Default value was not provided while defining the function.

~ The value of b is 2 because the default value of b = 6 will get override as we have assigned 2 for second argument (b) during function call. 

~ The value of c is 8 same as default value as we didn't override it in function call.

Question no.2 :- What is the result of this code, and why?

```python
def func(a, b, c=5):
    print(a, b, c)
func(1, c=3, b=2)
```

Answer no.2

~ The value of a is 1,which is assigned to it during function call. 

~ The value of b is 2 as assigned during function call. 

The value of c is 3 as we override the default value c = 5 in function call to c = 3.

Question no.3 :- How about this code: what is its result, and why?

```python
def func(a, *pargs):
    print(a, pargs)
func(1, 2, 3)
```

Answer no.3

~ The value of a which is default argument is 1 as it is assigned during function call. 

~ (2,3) is a tuple formed by arbitrary positional arguments.(*) asterisk is placed before any parameter in function definition which can hold n number of arguments and returns them in form of tuple.


Question no.4 :- What does this code print, and why?

```python
def func(a, **kargs):
    print(a, kargs)
func(a=1, c=3, b=2)
```

Answer no.4

~ The value of a which is default argument is 1 as assigned at the time of function call. 

~ {'c':3,'b':2} is the dictionary formed by arbitrary keyword arguments.(**)double asterisks are places before any parameter in function definition which can hold n number of keyword arguments and returns them in form of dictionary.


Question no.5 :-What gets printed by this, and explain?

```python
def func(a, b, c=8, d=5):
    print(a, b, c, d)
func(1, *(5, 6))
```

Answer no.5

~ The value of a is 1,which is assigned to it during function call.

~ The values of b and c are 5 and 6 respectively because (*) asterisk in function call will allow to add non keyword arguments. Even if keyword arguments are present by default it will override them. 

~ The value of d will be 5 as it is assigned in definition of function.
