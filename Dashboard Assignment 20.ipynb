Question no.4 :- Use the sqlite3 module to create a SQLite database called books.db, and a table called books with these fields: title (text), author (text), and year (integer).

Answer no.4

```python
import sqlite3
conn = sqlite3.connect('Booksdb.db')
cursor = conn.cursor()
sql = '''create table Books(
    title char(50), 
    author char(50), 
    year int)'''
cursor.execute(sql)
print("Table has been created") 
conn.commit()
```

Question no.5 :- Read books.csv and insert its data into the book table.

Answer no.5

```python
import pandas as pd
users = pd.read_csv('books.csv')
users.to_sql('Books',conn,if_exists='append',index=False)
```

Question no.6 :- Select and print the title column from the book table in alphabetical order.

Answer no.6

```python
cursor.execute('''select title from Books order by title asc''')
xf = pd.DataFrame(cursor.fetchall(),columns = ['title'])
xf
```

Question no.7 :- From the book table, select and print all columns in the order of publication.

Answer no.7

```python
cursor.execute('''select * from Books group by year order by year asc''')
vf = pd.DataFrame(cursor.fetchall(), columns = ['title','author','year'])
vf
```

Question no.8 :- Use the sqlalchemy module to connect to the sqlite3 database books.db that you just made in exercise 6.

Answer no.8

```python
import sqlalchemy
db_Engine = sqlalchemy.create_engine('sqlite:///Booksdb.db')
rows = db_Engine.execute('select * from Books')
for i in rows:
    print(i)
```
