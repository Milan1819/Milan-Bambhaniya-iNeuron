Question no.1 :- Describe three applications for exception processing.

Answer no.1

~ To Avoid programs or system from crashing.

~ To raise exceptions as and when required.

~ To Catch all the errors and log them into a file for further references.

Question no.2 :- What happens if you don't do something extra to treat an exception?

Answer no.2

~ If exception is not treated properly then the program stops and crashes, thus also the rest of the codes after that particular code with error will not execute.

Question no.3 :- What are your options for recovering from an exception in your script?

Answer no.3

~ Using exception clause with an exception message and we can also use exception in the else block of any script.

Question no.4 :- Describe two methods for triggering exceptions in your script.

Answer no.4

~ Assert statement and Raise statement are two ways for triggering exceptions.

Question no.5 :- Identify two methods for specifying actions to be executed at termination time, regardless of whether or not an exception exists.

Answer no.5

~ Finally block and Raise statement are actions which will be executed regardless of whether or not exception exists.
