Question no.5 :- Create a programme that searches a folder tree for files with a certain file extension(such as.pdf or.jpg). Copy these files from whatever location they are into a new folder.

Answer no.5

```python
import os,shutil

source_path = input("Enter the absolute path of the source folder: ")
file_type = input("Enter the type of file to copy (such as .pdf or .jpg): ").lower()
dest_path = input("Enter the absolute path of the destination folder: ")

for foldername, subfolders, filenames in os.walk(source_path):
    for filename in filenames:
        if filename.lower().endswith(file_type):
            print(f"{foldername} + ‘\\’ + {filename}")
            copy_source_path = os.path.join(foldername, filename)
            print(copy_source_path) 
            shutil.copy(copy_source_path, dest_path)
        else:
            continue
```
