Answer no.3

Question no.4 :- Where and how are instance attributes created?

Answer no.4

Question no.5 :- What does the term "self" in a Python class mean?

Answer no.5

~ The term "self" in Python class refers to as a pointer or references for all the instances, methods or attributes defined in the class to connect to the class.
