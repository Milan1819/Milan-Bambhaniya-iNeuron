Question no.1 :- What is the difference between __getattr__ and __getattribute__?

Answer no.1

~ __getattr__ : Is executed as the last resource when attribute is not found in an object.

~ __getattribute__ : Is used to retrieve an attribute from an instance. It captures every attempt to access an instance attribute by using dot notation or getattr() built-in function.

Question no.2 :- What is the difference between properties and descriptors?

Answer no.2

~ Descriptors are a low-level mechanism that lets you hook into an object's attributes being accessed. The Properties are a high-level application of this. Properties are implemented using descriptors.

Question no.3 :- What are the key differences in functionality between __getattr__ and __getattribute__, as well as properties and descriptors?

Answer no.3

~ The main difference between __getattr__ and __getattribute__ is that if the attribute was not found by the usual way then __getattr__ is used. Whereas the __getattribute__ is used before looking at the actual attributes on the object.

~ Descriptors are objects with ability to create objects that have special behavior when they're accessed as attributes of other objects whereas properties are implementation of descriptors.
