Question no.1 :- What is the purpose of Python's OOP?

Answer no.1

~ Object-oriented programming provides a means of structuring programs so that properties and behaviors are bundled into individual objects. It increases the reusability of codes and also makes coding less error prone.

Question no.2 :- Where does an inheritance search look for an attribute?

Answer no.2

~ An inheritance search looks for an attribute first in the instance object, then in the class from which the instance was created and then in all higher superclasses, progressing from left to right by default.

Question no.3 :-How do you distinguish between a class object and an instance object?

Answer no.3

~ In OOP, variables at the class level are referred to as class objects, whereas variables at the object level are called instance objects.
