Question no.5 :- Create a plain dictionary with the key-value pairs 'a': 1, 'b': 2, and 'c': 3, and print it out.

Answer no.5

```python
plain = dict() 
plain['a'] = 1
plain['b'] = 2
plain['c'] = 3

print(plain) 

for key,value in plain.items():
    print(key,value)
```

Question no.6 :- Make an OrderedDict called fancy from the same pairs listed in 5 and print it. Did it print in the same order as plain?

Answer no.6

```python
from collections import OrderedDict

fancy = OrderedDict()
fancy['a']= 1
fancy['b']= 2
fancy['c']= 3

print(fancy) 

for key,value in fancy.items():
    print(key,value)
```

~ OrderedDict is a subclass of dictionary class which maintains the elements in the inserted order.Normally Dictionary does not preserve the inserted order.

Question no.7 :- Make a default dictionary called dict_of_lists and pass it the argument list. Make the list dict_of_lists['a'] and append the value 'something for a' to it in one assignment. Print dict_of_lists['a'].

Answer no.7

```python
from collections import defaultdict
dict_of_lists = defaultdict()
dict_of_lists['a'] = 'something for a'
print(dict_of_lists['a'])
```
