Question no.1 :- Write a Python program to find LCM?

Answer no.1

```python
def get_lcm(x, y):

   if x > y:
       greater = x
   else:
       greater = y

   while(True):
       if((greater % x == 0) and (greater % y == 0)):
           lcm = greater
           break
       greater += 1

   return lcm

print("The L.C.M. is", get_lcm(50,30))
```

Question no.2 :- Write a Python program to find HCF?

Answer no.2

```python
def get_hcf(v,k):

    if v > k:
        small = k
    else:
        small = v
    for i in range(1, small+1):
        if((v % i == 0) and (k % i == 0)):
            hcf = i 
    return hcf

print("The H.C.F. is", get_hcf(64,44))
```

Question no.3 :- Write a Python program to convert Decimal to Binary, Octa and Hexadecimal?

Answer no.3

```python
def converter(d): 

    print(f"The decimal value of {d} is:")
    print(bin(d), "in binary.")
    print(oct(d), "in octal.")
    print(hex(d), "in hexadecimal.")

converter(631)
```

Question no.4 :- Write a Python program to find ASCII value of a character?

Answer no.4

```python
def ascii(character):
    return ord(character)

value = 'm'

M = ascii(value)

print(f"The ASCII value of {value} is: {M}")
```
