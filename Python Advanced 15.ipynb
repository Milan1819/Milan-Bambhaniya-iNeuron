Question no.2 :- What is monkey patching in Python?

Answer no.2

~ In Python, the term monkey patching refers to dynamic modifications of a class or module at run-time.

Question no.3 :- What is the difference between a shallow copy and deep copy?

Answer no.3

~ A shallow copy constructs a new compound object and then inserts references into it to the objects found in the original whereas a deep copy constructs a new compound object and inserts copies into it of the objects found in the original.

Answer no.4

Question no.5 :- What is generator comprehension?

Answer no.5

~ Python provides a sleek syntax for defining a simple generator in a single line of code, this expression is known as a generator comprehension.This produces a generator, whose instructions for generating its members are provided within the parenthetical statement.
