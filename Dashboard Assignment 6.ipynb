Question no.1 :- What are escape characters and how do you us them?

Answer no.1

~ Escape characters represent characters in string values that would otherwise be difficult or impossible to type into code.

~ An escape character is a backlash \ followed by the character you want to insert. e.g. \n, \r, \t etc.

Question no.2 :- What do the escape characters n and t stand for?

Answer no.2

~ \n stands for newline and \t stands for tab.

Question no.3 :- What is the way to include backslash characters in a string?

Answer no.3

~ \\ escape character can be used to represent backslash character in a string.

Question no.4 :- The string "Howl's Moving Castle" is a correct value. Why isn't the single quote character in the word Howl's not escaped a problem?

Answer no.4

~ It's not a problem because we have used double quotes at the beginning and end of the whole string so single quote in Howl's will work.

Question no.5 :- How do you write a string of newlines if you don't want to use the n character?

Answer no.5

~ By using triple quotes(''' ''') we can write string of newlines without using \n.

Question no.6 :- What are the values of the given expressions?

```python
a = 'Hello,world!'[1]
b = 'Hello,world!'[0:5]
c = 'Hello,world!'[:5]
d = 'Hello,world!'[3:]
```

Answer no.6

```python
print(f"The values of given expressions are {a} , {b} , {c} and {d} respectively")
```

Question no.7 :- What are the values of the following expressions?

```python
r = 'Hello'.upper()
s = 'Hello'.upper().isupper()
t = 'Hello'.lower()
```
