Question no.3 :- In a simple match of a string, which looks only for one match and does not do any replacement, is the use of a nontagged group likely to make any practical difference?

Answer no.3

~ In this Case The Non Tagged Group will not make any diffrence in this case.

```python
import re
phoneNumRegex = re.compile(r'\d\d\d')
num = phoneNumRegex.search('My number is 234-567-8901.')
print(f'Phone number found -> {num.group()}') # Non Tagged group
print(f'Phone number found -> {num.group(0)}') # Tagged Group
```

Question no.5 :- Unlike a normal regex pattern, a look-ahead condition does not consume the characters it examines. Describe a situation in which this could make a difference in the results of your programme.

Answer no.5

```python
example1 = re.search(r'checks(?=[a-z])',
                     "checksandchecks")
  
print('Using lookahead:', example1.group())
  

example2 = re.search(r'checks([a-z])',
                     "checksandchecks")
  
print('Without using lookahead:', example2.group())
```

~ Here, using lookahead the output generated is ‘checks’ whereas without using lookahead the output generated is checksa. The a is consumed by regex and it becomes part of the search string.

Question no.6 :- In standard expressions, what is the difference between positive look-ahead and negative look-ahead?

Answer no.6

~ Positive lookahead matches if the pattern matches what comes after the current location in the input string whereas negative lookahead matches if pattern does not match what comes after the current location in the input string.

Question no.8 :- Can you identify repeated items within a target string using named groups, as in "The cow jumped over the moon"?

Answer no.8

```python
match = re.search('(?P<name1>.*) (?P<name2>.*) (?P<name3>.*) (?P<name4>.*) (?P<name5>.*) (?P<name6>.*)','The cow jumped over the moon')
print(match.group('name1')) 
print(match.group('name5'))
```
