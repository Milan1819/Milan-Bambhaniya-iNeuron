Question no.1 :- Write a Python program to find sum of elements in list?

Answer no.1

```python
list = [7,8,6,4,6,8,9,7,2,1]

sum = 0

for i in list:
    sum = sum + i

print(f"The sum of elements in list is: {sum}")
```

Question no.2 :- Write a Python program to  Multiply all numbers in the list?

Answer no.2

```python
m = [6,2,3,8,9,5,5]

mul = 1

for i in m:
    mul = mul*i

print(f"The multiplication of numbers is list is: {mul}")
```

Question no.5 :- Write a Python program to find second largest number in a list?

Answer no.5

```python
sl = [22,35,56,47,80,98,74,21]

sl_new = set(sl)

sl_new.remove(max(sl_new))

print(max(sl_new),"is the second largest")
```

Question no.6 :- Write a Python program to find N largest elements from a list?

Answer no.6

```python
nl = [124,167,154,188,155,116,149,173]

n = int(input("Enter how many larger numbers you want: "))

nl.sort()

print(nl[-n:],"are",n,"largest elements from the list.")
```

Question no.7 :- Write a Python program to print even numbers in a list?

Answer no.7

```python
e = [2,3,5,7,11,19,18,17,6,4,22,33]

even_nums = []

for i in e:
    if i%2 == 0:
        even_nums.append(i)
        
print(even_nums)
```

Question no.8 :- Write a Python program to print odd numbers in a List?

Answer no.8

```python
o = [2,3,5,6,7,19,18,17,33,12,16]

odd_nums = []

for i in o:
    if i%2 != 0:
        odd_nums.append(i)

print(odd_nums)
```

Question no.9 :- Write a Python program to Remove empty List from List?

Answer no.9

```python
List = [1,2,4,[],[],7,8,9,[],11,19]

while [] in List:
    List.remove([])

print(List,"is list after removing empty lists.")
```

Question no.10 :- Write a Python program to Cloning or Copying a list?

Answer no.10

```python
def clone_spell(c):
    c_copy = []
    c_copy.extend(c) 
    return c_copy

IPL = ["RCB","CSK","GT","MI","KKR"]
IPL2 = clone_spell(IPL)

print("Original list:", IPL) 
print("Cloned list  :", IPL2)
```

Question no.11 :- Write a Python program to Count occurrences of an element in a list?

Answer no.11

```python
L = [3,4,5,5,6,7,8,8,8,8,9,9,9,2,7,1,8]

for i in range(len(L)):
    j = L.count(i)
    if j >= 1:
        print(f"{i} occurs {j} times.")
```
