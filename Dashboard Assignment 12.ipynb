Question no.1 :- In what modes should the PdfFileReader() and PdfFileWriter() File objects will be opened?

Answer no.1

~ PdfFileReader() and PdfFileWriter() should be opened in rb(read binary mode)  and wb(write binary mode)modes respectively.

Question no.2 :- From a PdfFileReader object, how do you get a Page object for page 5?

Answer no.2

~ PdfFileReader class has a method called getPage() which takes argument as page index number to get Page object. For calling Page object for page 5 we need to call getPage(4) as page 0 will be first page

Question no.6 :- What is the difference between a Run object and a Paragraph object?

Answer no.6

 Document object represents entire document. The document object contains list of Paragraph objects for the paragraphs in the document whereas each of these Paragraph objects contain list of one or more Run objects.

Question no.7 :- How do you obtain a list of Paragraph objects for a Document object that’s stored in a variable named doc?

Answer no.7

```python
from docx import Document

doc = Document("sample_file.docx")

print(doc.paragraphs)

for i in doc.paragraphs
    print(i.text)
```

Question no.10 :- How do you create a Document object for a new Word document?

Answer no.10

```python
first_doc = Document()

first_doc.add_paragraph("RCB won IPL 2023")

first_doc.save('myfirstdocument.docx')
```

Question no.11 :- How do you add a paragraph with the text 'Hello, there!' to a Document object stored in a variable named doc?

```python
doc = Document()

doc.add_paragraph('Hello,there!')

doc.save('hello.docx')
```
