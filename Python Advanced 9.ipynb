Question no.2 :- How do the string forms in Python 3.X vary in terms of operations?

Answer no.2

~ f-string: Letter “f” is placed before the beginning of the string, and the variables mentioned in curly braces will refer to the variables declared above.

~ format() method: format() method is called on a string object. Inside the string, we use curly braces {} that will refer to the format() method arguments. Number of {} should match the number of arguments inside format(). 

~ % operator: “%” operator will be replaced by variables defined in parenthesis/in tuple. %s means a string variable will come to this place, %d is an integer, %f is a floating-point value.
