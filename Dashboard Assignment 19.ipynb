Question no.8 :- Modify Element to make the attributes name, symbol, and number private. Define a getter property for each to return its value.

Answer no.8

```python
class Element:
    def __init__(self,name,symbol,number):
        self.__name = name 
        self.__symbol = symbol
        self.__number = number
    
    def name(self):
        return self.__name
    
    def symbol(self):
        return self.__symbol
    
    def number(self):
        return self.__number
```

```python
hydrogen3 = Element('Hydrogen','H',1)
print(hydrogen3.name()) 
print(hydrogen3.symbol())
print(hydrogen3.number())
```

Question no.10 :- Define these classes: Laser, Claw, and SmartPhone. Each has only one method: does(). This returns 'disintegrate' (Laser), 'crush' (Claw), or 'ring' (SmartPhone). Then, define the class Robot that has one instance (object) of each of these. Define a does() method for the Robot that prints what its component objects do.

Answer no.10

```python
class Laser:
    def does(self):
        return 'disintegrate'

class Claw:
    def does(self):
        return 'crush'

class Smartphone:
    def does(self):
        return 'ring'

class Robot:
    def __init__(self):
        self.laser = Laser().does() 
        self.claw = Claw().does() 
        self.smartphone = Smartphone().does() 

    def does(self):
        print(f"Robot has laser to: {self.laser}, claws to: {self.claw} and smartphone to: {self.smartphone}")
```

```python
R = Robot()
print(R.does())
```
