Question no.1 :- What is the relationship between def statements and lambda expressions?

Answer no.1

~ Both are used create functions.A def is a statement to create function and lambda is a expression to create function.

Question no.2 :- What is the benefit of lambda?

Answer no.2

~ Lambda helps to use a function only once, and hence, avoids cluttering up the code with function definitions. It lets you define a function in a single line of code and use it immediately.

Question no.3 :- Compare and contrast map, filter, and reduce.

Answer no.3

~ All the three functions take a function and iterable as arguments, the function is applied to iterable to get results. However,the functionality and outputs of all the three functions will be different.

~ map() applies function to all the elements of arguments.

~ filter() filters out given arguments as per the outcome of the function based on True False condition.

~ reduce() applies the function to the elements of the sequence, from left to right, starting with the first two elements in the sequence. It reduces the sequence to single value as output.


Question no.4 :- What are function annotations, and how are they used?

Answer no.4

~ Function annotations are some random expressions which are written with the functions, and they are evaluated at compile time. They do not exist at run time, and there is no meaning of these expressions to python. They are used and interpreted by a third party or external python libraries.

Question no.5 :- What are recursive functions, and how are they used?

Answer no.5

~ Recursive functions are functions that call itself. It is always made up of 2 conditions, the base case and the recursive case. The base case is the condition to stop the recursion. The recursive case is the condition where the function calls on itself.Recursion is a technique used to solve computer problems by creating a function that calls itself until your program achieves the desired result.
