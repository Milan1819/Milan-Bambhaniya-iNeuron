Question no.1 :- Write a Python program to display Fibonacci sequence using Recursion?

Answer no.1

```python
def recursion(n):
   if n <= 1:
       return n
   else:
       return(recursion(n-1) + recursion(n-2))

t = int(input("Enter how many no. of sequence you want: ")) 

if t <= 0:
   print("Plese input positive integer")

else:
   print("Fibonacci sequence is:")
   for i in range(t):
       print(recursion(i))
```

Question no.2 :- Write a Python program to find Factorial of a number using Recursion?

Answer no.2

```python
def recur(a):
   if a == 1:
       return a
   else:
       return a*recur(a-1)

b = int(input("Enter number to get it's factorial: ")) 

if b < 0:
   print("Factorial doesn't exist for negative numbers")

elif b == 0:
   print("The factorial of 0 is 1")

else:
   c = recur(b) 
   print(f"The factorial of {b} is: {c}")
```

Question no.3 :- Write a Python program to find your Body Mass Index?

Answer no.3

```python
def Body_Mass_Index(height,weight):
    bmi = weight/(height**2) 
    return bmi

w = float(input("Enter your weight in kg: "))
h = float(input("Enter your height in metres: "))

I = Body_Mass_Index(h,w)
print(f"Your Body Mass Index is: {I}")
```

Question no.5 :- Write a Python program for cube sum of first n natural numbers?

Answer no.5

```python
def cube_sum(q):

   sum = 0
   for i in range(1, q+1):
      sum +=i*i*i
   return sum

q = int(input("Enter value to get cube sum upto it: ")) 
Q = cube_sum(q)
print(f"The Cube Sum of {q} natural numbers is: {Q}")
```
