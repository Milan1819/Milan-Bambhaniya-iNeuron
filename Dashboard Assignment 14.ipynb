Question no.4 :- Use your image and load in notebook then, How can you find out the width and height of an Image object?

Answer no.4

~ The following code can be used to get height and width of an image object:

```python

from PIL import Image

image_file = 'any_image.jpg' # add any image filepath to be checked
img = Image.open(image_file)
w = img.width 
h = img.height

print(f"The width and height of an image object are {w} and {h} respectively")
```
