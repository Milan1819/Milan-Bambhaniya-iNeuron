Question no.1 :- Write a Python program to Extract Unique values from dictionary values?

Answer no.1

```python
my_dict = {'A' : [7, 3, 7, 8],
             'B' : [9, 16, 11, 15],
             'C' : [7, 12, 22 ,7],
             'D' : [19, 15, 8]}
  
u_values = []

for i in my_dict.values():
    u_values.extend(i) 

u_values = list(set(u_values))

print("The unique values are:",u_values)
```

Question no.4 :- Write a Python program to convert key-values list to flat dictionary?

Answer no.4

```python
from itertools import product

my_dict2 = {"teams":["RCB","MI","DC","KKR"], 
            "points_table":[1,2,3,4]}

flat_dict = dict(zip(my_dict2["teams"],my_dict2["points_table"]))

print("The Flattened Dictionary is:")
print(flat_dict)
```

Question no.5 :- Write a Python program for insertion at the beginning in OrderedDict?

Answer no.5

```python
from collections import OrderedDict
 
my_dict3 = OrderedDict([('x',5),('y',8)]) 
my_dict3.update({'z':3}) 
my_dict3.move_to_end('z',last = False)

print(my_dict3)
```

Question no.6 :- Write a Python program to check order of character in string using OrderedDict()?

Answer no.6

```python
def check_order(input,pattern):
   my_dict4 = OrderedDict.fromkeys(input)
   
   length = 0
   for key,value in my_dict4.items():
      if (key == pattern[length]):
         length = length + 1

      if (length == (len(pattern))):
         return 'The order of pattern is correct'

   return 'The order of pattern is incorrect'

input = 'Abraham Benjamin de Villiers'
pattern = 'ABd'  
print("The string is: ")
print(input)
print("The pattern is: ")
print(pattern)
print(check_order(input,pattern))
```
