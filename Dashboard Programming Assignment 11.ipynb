Question no.1 :- Write a Python program to find words which are greater than given length k?

Answer no.1

```python
W = "The pitch of the Indore Test match was rated poor by ICC"

k = int(input("Enter length of word: "))
 
new_W = []
for word in W.split():
    if len(word) > k:
        new_W.append(word)

print(new_W)
```

Question no.4 :- Write a Python to check if a given string is binary string or not?

Answer no.4

```python
S = input("Enter the string to check: ")

if S.count("0") + S.count("1") == len(S):
    print("It is binary string.")
else:
    print("It is not a binary string")
```

Question no.5 :- Write a Python program to find uncommon words from two Strings?

Answer no.5

```python
def Uncommon(A,B):
    A = A.split()
    B = B.split()
    L = []

    for i in A:
        if i not in B:
            L.append(i)

    for i in B:
        if i not in A:
            L.append(i)
    L = set(L)
    return L

A = "Batman Superman Flash Vigilante"
B = "Joker Peacemaker Superman Flash"

print(Uncommon(A,B))
```

Question no.6 :- Write a Python program to find all duplicate characters in string?

Answer no.6

```python
def Duplicates(M):
    x=[]

    for i in M:
        if i not in x and M.count(i)>1:
            x.append(i)
    print(" ".join(x))
 


M = input("Enter a string: ") 
Duplicates(M)
```

Question no.7 :- Write a Python program to check if a string contains any special character?

Answer no.7

```python
E = input("Enter the string to check: ") 
E.split()

flag = False

sp_c = '[@_!#$%^&*()<>?/\|}{~:]'

for i in range(len(E)):
    if E[i] in sp_c:
        flag = True     

if flag:
    print("String contains special character") 

else:
    print("String doesn't have special character")
```
