Question no.4 :- What is the most effective way to initialise a list of 99 integers that repeats the sequence 1, 2, 3? S If necessary, show step-by-step instructions on how to accomplish this.

Answer no.4

```python
rn = []
for i in range(1,100):
    for num in range(1,4):
        rn.append(num)
print(rn,end = " ")
```
