Question no.1 :-

Please write a program using generator to print the numbers which can be divisible by 5 and 7 between 0 and n in comma separated form while n is input by console.

Example:
If the following n is given as input to theprogram:
100

Then, the output of the program should be:
0,35,70

Answer no.1

```python
def generator(n):
    
    for i in range(0,n+1):
        if i%5 == 0 and i%7 == 0:
            yield i
    

n = int(input("Enter the range: ")) 

vals = []

for i in generator(n):
    vals.append(str(i)) 

print(",".join(vals))
```

Question no.2 :-

Please write a program using generator to print the even numbers between 0 and n in comma separated form while n is input by console.

Example:
If the following n is given as input to the program:
10

Then, the output of the program should be:
0,2,4,6,8,10

Answer no.2

```python
def Even(n):

    for i in range(0,n+1):
        if i%2 == 0:
            yield i

n = int(input("Enter the range: ")) 

even_nums = []

for i in Even(n):
    even_nums.append(str(i))

print(",".join(even_nums))
```

Question no.3 :-

The Fibonacci Sequence is computed based on the following formula:

f(n)=0 if n=0

f(n)=1 if n=1

f(n)=f(n-1)+f(n-2) if n>1

Please write a program using list comprehension to print the Fibonacci Sequence in comma separated form with a given n input by console.

Example:
If the following n is given as input to the program:7

Then, the output of the program should be:
0,1,1,2,3,5,8,13

Answer no.3

```python
n = int(input("Enter the input: "))
f = lambda x: 0 if x == 0 else 1 if x == 1 else f(x-1)+f(x-2)
print(','.join([str(f(x)) for x in range(0, n+1)]))
```

Question no.4 :-

Assuming that we have some email addresses in the "username@companyname.com" format, please write program to print the user name of a given email address. Both user names and company names are composed of letters only.

Example:

If the following email address is given as input to the program:

john@google.com

Then, the output of the program should be:

john

Answer no.4

```python
import re
email = input("Enter email address: ")
name = "(\w+)@(\w+)\.(com)"
N = re.match(name,email)
print(N.group(1))
```

Question no.5 :-

Define a class named Shape and its subclass Square. The Square class has an init function which takes a length as argument. Both classes have a area function which can print the area of the shape where Shape's area is 0 by default.

Answer no.5

```python
class Shape():
    def __init__(self):
        pass

    def area(self):
        return 0

class Square(Shape):
    def __init__(self,length):
        Shape.__init__(self)
        self.length = length

    def area(self):
        return self.length**2

length = int(input("Enter length of square: "))

shape_o = Shape() 
square_o = Square(length)
print("The Area of shape : ",shape_o.area()) 
print("The Area of Square: ",square_o.area())
```
