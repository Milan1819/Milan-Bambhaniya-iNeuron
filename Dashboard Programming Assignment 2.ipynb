Question no.1 :- Write a Python program to convert kilometers to miles?

Answer no.1

```python
km = float(input("Enter kilometers to convert: "))
miles = 0.621371*km
print(f"The miles for given kilometers is: {miles}")
```

Question no.2 :- Write a Python program to convert Celcius to Fahrenheit?

Answer no.2

```python
celcius = float(input("Enter Celcius value for conversion: "))
fahrenheit = (celcius*1.8)+32
print(f"{celcius} degree Celcius is equal to degree {fahrenheit} Fahrenheit")
```

Question no.3 :- Write Python program to display calendar?

Answer no.3

```python
import calendar
year = int(input("Enter Year: "))
month = int(input(" Enter Month: "))
print(calendar.month(year,month))
```

Question no.4 :- Write a Python program to solve quadratic equation?

Answer no.4

```python
import cmath

a = 3
b = 7
c = 9

d = (b**2) - (4*a*c)

r1 = (-b-cmath.sqrt(d))/(2*a)
r2 = (-b+cmath.sqrt(d))/(2*a)

print("The solutions are: ")
print(r1) 
print(r2)
```

Question no.5 :- Write a Python program to swap two variables without temp variable?

Answer no.5

```python
x = float(input("Enter number x: "))
y = float(input("Enter number y: "))

x,y = y,x

print(f"The swapped value of x is: {x}") 
print(f"The swapped value of y is: {y}")
```
