Question no.2 :- Write a Python program to do arithmetical operations addition and division.?

Answer no.2

Question no.3 :- Write a Python program to find the area of a triangle?

Answer no.3

```python
h = float(input('Enter first side: '))
k = float(input('Enter second side: '))
j = float(input('Enter third side: '))

s = (h + k + j) / 2

area = (s*(s-h)*(s-k)*(s-j)) ** 0.5
print(f"The Area of the triangle is : {area}")
```

Question no.4 :- Write a Python program to swap two variables?

Answer no.4

```python
x = int(input("Enter first Integer X : "))
y = int(input("Enter second Integer Y: "))

z = x
x = y
y = z

print(f"The value of x after swapping is: {x}")
print(f"The value of y after swapping is: {y}")
```
