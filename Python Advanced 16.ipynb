Question no.6 :- What are some examples of how to use the match object returned by re.match and re.search?

Answer no.6

```python
import re
 
Substring ='RCB'
 
String1 = "RCB will win IPL 2023 this time"

String2 = "CSK will win IPL 2023 this time"
 

print(re.search(Substring, String1, re.IGNORECASE))

print(re.match(Substring, String1, re.IGNORECASE))

print(re.search(Substring, String2, re.IGNORECASE))

print(re.match(Substring, String2, re.IGNORECASE))
```
