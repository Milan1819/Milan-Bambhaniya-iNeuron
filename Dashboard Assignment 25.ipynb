Question no.1 :- What is the difference between enclosing a list comprehension in square brackets and parentheses?

Answer no.1

~ The list comprehension enclosed in square brackets produces an output in the form of list. 

~ The list comprehension enclosed in parathesis produces a generator object. The generator object can be iterated to yield items one by one in the output.

Question no.2 :- What is the relationship between generators and iterators?

Answer no.2

~ The generators are object that support iteration protocol. The generators have a next method which repeatedly advances to next item in a series of results.

Question no.3 :- What are the signs that a function is a generator function?

Answer no.3

~ A generator function has a yield statement somewhere in it's code which signs that it is a generator function.

Question no.4 :- What is the purpose of a yield statement?

Answer no.4

~ When present, yield statement makes python compile the function specially as a generator. When called it returns a generator object that supports the iteration protocol. When the yield statement runs it produces results one by one from the series of results.

Question no.5 :- What is the relationship between map calls and list comprehensions? Make a comparison and contrast between the two.

Answer no.5

~ List comprehension returns a list, whereas the map function returns an object of Iterable. List comprehension execution is faster than that of map function when the formula expression is huge and complex. Map function is faster than list comprehension when the formula is already defined as a function earlier.
