Question no.1 :- What is the concept of an abstract superclass?

Answer no.1

~ An abstract superclass in Python is to declare a set of methods that must be created in any child class built on top of this abstract class. It is used for hiding unnecessary information from the users.

Question no.2 :- What happens when a class statement's top level contains a basic assignment statement?

Answer no.2

~ It will throw error called Nameerror as class uses __init__ constructor to define it's attributes and not by assignment statement.

Question no.3 :- Why does a class need to manually call a superclass's init method?

Answer no.3

~ The main reason for always calling superclass's __init__ method is that superclass may typically create member variable and initialize them to defaults. So if we don't call superclass __init__, none of that code would be executed and you would end up with superclass that has no member variables.

Question no.4 :- How can you augment, instead of completely replacing, an inherited method?

Answer no.4

~ Function Overriding can be used to augment instead of replacing an inherited method. We need to create a function in child class with same name and number of parameters.

Question no.5 :- How is the local scope of a class different from that of a function?

Answer no.5

~ When a class is defined, a new namespace is created, and used as the local scope, so all assignments to local variables go into this new namespace.
