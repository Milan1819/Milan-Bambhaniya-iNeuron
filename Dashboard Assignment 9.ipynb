Question no.1 :- To what does a relative path refer?

Answer no.1

~ Relative paths refer to current working directory.

Question no.2 :- What does an absolute path start with your operating system?

Answer no.2

~ Absolute paths start with slash(/) or C:/.

Question no.3 :- What do the functions os.getcwd() and os.chdir() do?

Answer no.3

~ os.getcwd() returns the current working directory and os.chdir() changes the current working directory.

Question no.4 :- What are the . And .. folders?

Answer no.4

~ . folder is current folder and .. folder is the parent folder.

Question no.5 :- In C:\bacon\eggs\spam.txt, which part is the dir name and which part is the base name?

Answer no.5

~ C:\bacon\eggs is the dir name and spam. txt is the base name.

Question no.6 :- What are the three "mode" arguments that can be passed to the open() function?

Answer no.6

~ 'r' for read mode, 'w' for write mode and 'a' for append mode.

Question no.7 :- What happens if an existing file is opened in write mode?

Answer no.7

~ If an existing file is opened in write mode then it gets erased and completely overwritten.

Question no.8 :- How do you tell the difference between read() and readlines()?

Answer no.8

~ The read() method returns the file's entire contents as a single string value and the readlines() method returns a list of strings, where each string is a line from the file's contents.

Question no.9 :- What data structure does a shelf value resemble?

Answer no.9

~ A shelf value resembles a dictionary value. It has keys and values along with keys() and values() methods which work same as dictionary methods with same name.
