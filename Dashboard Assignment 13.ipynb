Question no.5 :- What do the keyword arguments delimiter and line terminator do?

Answer no.5

~ Delimeter is a sequence of one or more characters used to specify boundary between separate data strings or code set characters. ,(comma) is an example.

~ A line terminator is used to terminate lines. It marks the end of a line of input character sequence. \n(newline) is an example.
