Question no.3 :- Write a Python program for array rotation?

Answer no.3

```python
def rotate_array(a,x):
    l=len(a)
    a[:]=a[x:l]+a[0:x]
    return a

arr = [1, 2, 3, 4, 5, 6]

print("Rotated array is")
print(rotate_array(arr,5))
```

Question no.5 :- Write a Python Program to check if given array is Monotonic?

Answer no.5

```python
def Monotonic(M):

    x = []
    y = []
    x.extend(M)
    y.extend(M)
    x.sort()
    y.sort(reverse=True)

    if(x == M or y == M):
        return True
    else:
        return False

M = [26, 15, 13, 10, 7, 3]

print(Monotonic(M))
```
