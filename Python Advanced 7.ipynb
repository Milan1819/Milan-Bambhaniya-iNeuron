Question no.1 :- What is the purpose of try statement?

Answer no.1

~ The purpose of try block is to check the block of code specified in it for errors.

Question no.2 :- What are the two most popular try statement variations?

Answer no.2

~ else and finally are two variations of try statement.

Question no.3 :- What is the purpose of the raise statement?

Answer no.3

~ The raise statement is used to raise an exception. You can define what kind of error to raise, and the text to print.

Question no.4 :- What does the assert statement do, and what other statement is it like?

Answer no.4

~ The assert statement lets you test if a condition in your code returns True, if not, the program will raise an AssertionError. It is similar to raise statement.

Question no.5 :- What is the purpose of the with/as argument, and what other statement is it like?

Answer no.5

~ The with/as argument in Python is used for resource management and exception handling. It is generally used when working with file streams. It is similar to try/finally statements.
