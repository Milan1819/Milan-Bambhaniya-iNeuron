Question no.7 :- Write a generator, genPrimes, that returns the sequence of prime numbers on successive calls to its next() method: 2, 3, 5, 7, 11, ...

Answer no.7

```python
from math import sqrt

def is_prime(n):
    if (n <= 1):
        return False
    if (n == 2):
        return True
    if (n % 2 == 0):
        return False

    i = 3
    while i <= sqrt(n):
        if n % i == 0:
            return False
        i = i + 2

    return True
```

```python
def genPrimes():
    n = 1
    while True:
        n += 1
        if is_prime(n):
            yield n
```

```python
result = genPrimes()

for i in range(10):
    print(next(result),end=' ')
```
