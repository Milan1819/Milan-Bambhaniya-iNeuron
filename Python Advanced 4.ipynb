Question no.1 :- Which two operator overloading methods can you use in your classes to support iteration?

Answer no.1

~ __iter__ and __next__ are the two operator overloading methods we can use in classes to support iteration.

Question no.3 :- In a class, how do you intercept slice operations?

Answer no.3

~ By using slice() method we can do slice operations in a class. The slice() method returns a portion of an iterable as an object of the slice class based on the given range.

Question no.4 :- In a class, how do you capture in-place addition?

Answer no.4

~ __iadd__() method is used to capture in_place addition in a class.

Question no.5 :- When is it appropriate to use operator overloading?

Answer no.5

~ When it is ensured that objects of a class behave consistently with built-in types and other user-defined types, we consider using operator overloading as appropriate.
