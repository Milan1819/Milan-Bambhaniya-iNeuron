Question no.1 :- What is the result of the code, and explain?

```python
X = 'iNeuron'
def func():
    print(X)


func()
```

Answer no.1

~ A variable defined outside any function is called global variable. Here, X is a global variable which can be called anywhere inside or outside of any function.

Question no.3 :- What does this code print, and why?

```python
X = 'iNeuron'
def func():
    X = 'NI'
    print(X)
   

func()
print(X)
```

Answer no.3

~ Here both values of variable X are printed. If you operate with the same variable name inside and outside of a function, Python will treat them as two separate variables, one available in the global scope (outside the function) and one available in the local scope (inside the function).

Question no.4 :- What output does this code produce? Why?

```python
X = 'iNeuron'
def func():
    global X
    X = 'NI'


func()
print(X)
```

Answer no.4

~ global keyword is used ahead of variable name to declare a global variable inside function and to change the value of global variable declared in global scope(outside of the function). Hence, NI is printed.

Question no.5 :- What about this code—what’s the output, and why?

```python
X = 'iNeuron'
def func():
    X = 'NI'
    def nested():
        print(X)
    nested()


func()
X
```

Answer no.5

~ Here NI is received in output by using the nested function or inner function. Nested functions is used to access variables of enclosing scope. Enclosing (or nonlocal) scope is a special scope that only exists for nested functions. If the local scope is an inner or nested function, then the enclosing scope is the scope of the outer or enclosing function. 

~ iNeuron is received in output through global variable X defined global scope.

Question no.6 :- How about this code: what is its output in Python 3, and explain?

```python
def func():
    X = 'NI'
    def nested():
        nonlocal X
        X = 'Spam'
    nested()
    print(X)


func()
```

Answer no.6

~ Here output is Spam because nonlocal keyword is used with variable X in nested function which is used to change the value of variable defined in outer function.
