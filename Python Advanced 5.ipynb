Question no.1 :- What is the meaning of multiple inheritance?

Answer no.1

~ Multiple Inheritance is a process where the child class or object inherits the methods and attributes from one or more parent classes.

Question no.2 :- What is the concept of delegation?

Answer no.2

~ Delegation is an object oriented technique. It refers to evaluating a member (property or method) of one object (the receiver) in the context of another original object (the sender).

Question no.3 :- What is the concept of composition?

Answer no.3

~ Composition is to enable creating complex types by combining objects of different classes. It means that a class Composite can contain an object of another class Component. This type of relationship is known as Has-A Relation.

Question no.4 :- What are bound methods and how do we use them?

Answer no.4

~ If a function is an attribute of class and it is accessed by the instances, they are called bound methods. They take an instance of their class along with the other arguments as its parameters.

Question no.5 :- What is the purpose of pseudoprivate attributes?

Answer no.5

~ Pseudoprivate attributes are useful in larger frameworks or tools, both to avoid introducing new method names that might accidentally hide definitions elsewhere in the class tree and to reduce the chance of internal methods being replaced by names defined lower in the tree.
