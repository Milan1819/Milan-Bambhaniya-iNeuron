Question no.1 :- Does assigning a value to a string's indexed character violate Python's string immutability?

Answer no.1

~ Strings in Python are immutable which means that once a string variable is assigned to a string the contents of the string cannot be changed unlike the list object.

Question no.2 :- Does using the += operator to concatenate strings violate Python's string immutability? Why or why not?

Answer no.2

~ It doesn't violate because the plus equal operator (+=) appends to strings and creates a new string while not changing the value of the original string.

Question no.3 :- In Python, how many different ways are there to index a character?

Answer no.3

Question no.4 :- What is the relationship between indexing and slicing?

Answer no.4

~ Indexing is used to obtain individual elements. Slicing is used to obtain a sequence of elements. Indexing and Slicing can be be done in Python Sequences types like list, string, tuple, range objects and more.

Question no.5 :- What is an indexed character's exact data type? What is the data form of a slicing-generated substring?

Answer no.5

~ Indexed character's data type is str and data form of slicing-generated substring will be same as the iterable on which slicing is performed. e.g. slicing on string will return datatype as str, slicing on list will return datatype as list.

Answer no.6
