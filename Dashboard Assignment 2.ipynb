Answer no.3

~ AND : True only if all conditions are true. For example x and y are true only if both of them are true.

~ OR : True if atleast one condition is true. For example x and y are true if either one of them is true.

~ NOT : True only if condition is False.

Question no.12 :- Write a short program that prints the numbers 1 to 10 using for loop. Then write an equivalent program that prints the numbers 1 to 10 using while loop.

Answer no.12

```python
# Using for loop

for i in range(1,11):
    print(i,end=" ")
```

```python
# Using while loop 

i=1
while i in range(1,11):
    print(i,end=" ")
    i=i+1
```
