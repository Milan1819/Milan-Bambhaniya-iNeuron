Question no.1 :- Is the Python Standard Library included with PyInputPlus?

Answer no.1

~ No, it is not included with Python Standard Library, we need to install it separately by pip install command.

Question no.2 :- Why is PyInputPlus commonly imported with import pyinputplus as pypi?

Answer no.2

~ Because we can use the short form(pypi) as alias while calling the module's functions which increases the ease of coding.

Question no. 3 :- How do you distinguish between inputInt() and inputFloat()?

Answer no.3

~ inputInt() is used for getting integer value and inputFloat() is used for getting float value.

Question no.4 :- Using PyInputPlus, how do you ensure that the user enters a whole number between 0 to 99?

Answer no.4

~ By using pyip.inputint(min=0,max=99) we can ensure that user enters a whole number between 0 to 99.

Question no.5 :- What is transferred to the keyword arguments allowRegexes and blockRegexes?

Answer no.5

~ A list of regex strings that are either explicitly allowed or denied respectively.

Question no.6 :- If a blank input is entered three times, what does inputStr(limit=3)do?

Answer no.6

~ It will raise RetryLimitException.

Question no.7 :- If blank input is entered three times, what inputStr(limit=3,default='hello')do?

Answer no.7

~ It returns the value 'hello'.
