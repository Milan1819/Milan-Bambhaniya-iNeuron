Question no.4 :- What are the two lines that must be present in your software in order to call logging.debug()?

Answer no.4

~ To be able to call logging.debug(), you must have these two lines at the start of your program:

```python
import logging 
logging.basicConfig(level=logging.DEBUG, format=' %(asctime)s - %(levelname)s - %(message)s')
```

Question no.5 :- What are the two lines that your program must have in order to have logging.debug() send a logging message to a file named programLog.txt?

Answer no.5

```python
import logging 
logging.basicConfig(filename='programLog.txt', level=logging.DEBUG, format=' %(asctime)s - %(levelname)s - %(message)s')
```

Question no.9 :- What are the differences between the Step Over, Step In, and Step Out buttons in the debugger?

Answer no.9

~ Step Over - Stepping over means to move to the next line of code in the current function. Even if the current line is a function call, all the code for that function will be executed and the new current next line will be the next line of the code that you were looking for.

~ Step In - Complete the next line of code. If this line contains a function, go to the first line of the function's code and stop, waiting for the user to now debug the function.

~ Step Out - If the current line of execution in the program is inside the code of a function and you want to finish it, you can use the step out command. This will complete all the code in the current function and return you to the previous function that "called" this function

Question no.10 :- After you click Continue, when will the debugger stop ?

Answer no.10

Question no.11 :- What is the concept of a breakpoint?

Answer no.11

~ A breakpoint is a point in the program where the code will stop executing. Breakpoint is to set a line of code that causes the debugger to pause when the program execution reaches that line.
